```python
import pandas as pd
```

```python
df=pd.read_csv("deliveries.csv")
df
```

```python
# checking the info of the dataset
df.info()
```

```python
# checking the unique values in Extras_type and filling the null values with "Normal"
df['extras_type'].unique()
df['extras_type']=df['extras_type'].fillna("Normal")
```

```python
# checking the unique values in player_dismissed and filling the null values with "Not Dismissed"
print(df['player_dismissed'].unique())
df['player_dismissed']=df['player_dismissed'].fillna("Not Dismissed")

# checking the unique values in fielder and filling the null values with "No Fielder"
print(df['fielder'].unique())
df['fielder']=df['fielder'].fillna("No Fielder")

# checking the unique values in dismissal_kind and filling the null values with "Not Out"
print(df['dismissal_kind'].unique())
df['dismissal_kind']=df['dismissal_kind'].fillna("Not Out")
```

```python
df.info()
```

```python
# Replace team names and city names for consistency
df['batting_team'] = df['batting_team'].replace('Delhi Daredevils', 'Delhi Capitals')
df['bowling_team'] = df['bowling_team'].replace('Delhi Daredevils', 'Delhi Capitals')
df['batting_team'] = df['batting_team'].replace('Deccan Chargers', 'Sunrisers Hyderabad')
df['bowling_team'] = df['bowling_team'].replace('Deccan Chargers', 'Sunrisers Hyderabad')
df['batting_team'] = df['batting_team'].replace('Rising Pune Supergiants', 'Rising Pune Supergiant')
df['bowling_team'] = df['bowling_team'].replace('Rising Pune Supergiants', 'Rising Pune Supergiant')
df['batting_team'] = df['batting_team'].replace('Royal Challengers Bangalore', 'Royal Challengers Bengaluru')
df['bowling_team'] = df['bowling_team'].replace('Royal Challengers Bangalore', 'Royal Challengers Bengaluru')
df['batting_team'] = df['batting_team'].replace('Kings XI Punjab', 'Punjab Kings')
df['bowling_team'] = df['bowling_team'].replace('Kings XI Punjab', 'Punjab Kings')

```

```python
df.to_csv("cleaned_deliveries.csv", index=False)
```
